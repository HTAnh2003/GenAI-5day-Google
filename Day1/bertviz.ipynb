```python
from bertviz import head_view
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch
```

```python
model_name = "SpanBERT/spanbert-base-cased"  # Or a BERT-based model suitable for co-reference
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
text = "My car is over there. It is just washed."
inputs = tokenizer(text, return_tensors="pt")
```

```python
outputs = model(**inputs, output_attentions=True)
attention = outputs.attentions  # This contains attention weights for all layers and heads
```

```python
head_view(attention, tokens=tokenizer.convert_ids_to_tokens(inputs.input_ids[0]))
```
